# Modelling Phase

## Training Phase

```python
# Machine learning model 1 training phase

X_train = data_train[['carat']]
y_train = data_train['price']

X_test = data_test[['carat']]
y_test = data_test['price']
```

```python
clf = LinearRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
# Machine learning model 2 training phase (multiple linear regression)

X2_train = data_train.drop(['price'], axis = 1)
y2_train = data_train['price']

X2_test = data_test.drop(['price'], axis = 1)
y2_test = data_test['price']
```

```python
clf2 = LinearRegression()
```

```python
clf2.fit(X2_train, y2_train)
```

## Testing Phase

```python
# Machine learning model 1 testing phase
pred = clf.predict(X_test)
pred
```

```python
pd.DataFrame({"Aktual":y_test,
             "Prediksi":pred})
```

```python
# Machine learning model 2 testing phase (multiple linear regression)

pred2 = clf2.predict(X2_test)
pred2
```
